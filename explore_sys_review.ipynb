```python
def check_word_fraction_in_df(df, word):
    fraction = len(df[df['Title'].str.contains(word) | df['Abstract'].str.contains(word)])/len(df)
    return fraction
```

```python
words_list = ['context','objectives', 'design', 'outcome measures', 'data source', 'results', 'conclusions']
dfs= [citation_included, citation_excluded]
for word in words_list:
    frac = check_word_fraction_in_df(citation_included, word)
    print(f"Fraction of samples with word [{word}] in included citations : {frac}")
    frac = check_word_fraction_in_df(citation_excluded, word)
    print(f"Fraction of samples with word [{word}] in excluded citations : {frac}")
    
```
